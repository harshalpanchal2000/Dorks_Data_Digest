```python
df= pd.read_csv("/kaggle/input/qqrq444/output.csv")
df.head(10)
```

```python
# Keywords for each cluster
clusters_keywords = {
    "Data Analysis": ["data analysis", "data visualization"],
    "Machine Learning": ["machine learning", "supervised learning", "unsupervised learning"],
    "Deep Learning": ["deep learning", "neural networks", "deep neural networks"],
    "Computer Vision": ["computer vision", "image processing", "object detection"],
    "NLP": ["natural language processing", "text processing", "word embeddings"],
    "AI": ["artificial intelligence", "reinforcement learning", "expert systems"],
    "Math": ["mathematics", "mathematical modeling", "mathematical methods", "statistics", "statistical analysis"]
}

# Initialize clusters as empty dictionaries
clusters = {topic: pd.DataFrame(columns=df.columns) for topic in clusters_keywords}

# Assign books to clusters based on keywords
for i, row in df.iterrows():
    title = row['title']
    for topic, keywords in clusters_keywords.items():
        for keyword in keywords:
            if keyword in title.lower():
                clusters[topic] = pd.concat([clusters[topic], row.to_frame().T], ignore_index=True)
                break

# Export each DataFrame to a CSV file
for topic, df_cluster in clusters.items():
    filename = f"{topic}.csv"
    df_cluster.to_csv(filename, index=False)
    print(f"{filename} exported successfully.")
```
